# Progetto Python: Analisi dei piloti mondiale Formula 1

In questo Notebook, a partire da un file CSV, analizziamo e gestiamo le principali statistiche dei piloti, e dei costruttori del mondiale di Formula 1 del 2008.

### Funzioni presenti nel Notebook:
1. Conversione(posizione):
Converte nei punteggi validi nel 2008 i posizionamenti ottenuti in gara.

2. Trova_pilota(lista_valori):
Chiede l'inserimento in input del nome di un pilota e, se presente, restituisce le statistiche del mondiale 2008.

3. Classifica_mondiale_piloti(lista_valori):
Conteggia i punti ottenuti da ongi singolo pilota presente e restituisce la classifica finale, ordinata, tramite dizionario.
Stampiamo il risultato su un file .txt chiamato "Drivers_Standings_2008".
(NB. La stampa avviene dopo aver chiamato la funzione, non all'interno della stessa).

4. Classifica_mondiale_costruttori(lista_valori):
Effettua il calcolo dei punti ottenuti dai singoli Team e restituisce la classifica finale, ordinata, su dizionario.




```python
import csv # Importo il modulo csv per la lettura del un file

# In un contesto apriamo il file e assegnamo ad una variabile il suo contenuto creando una lista di dizionari.
with open("/content/drive/MyDrive/Progetto Python/formula1_data.csv") as f:
  reader = csv.DictReader(f)
  dati=list(reader)
```

```python
"""
Funzioni
"""
def conversione(posizione):
  """
  La funzione converte i valori del posizionamento in classifica con il punteggio relativo al regolamento F1 2008:
  1°: 10 punti
  2°: 8 punti
  3°: 6 punti
  4°: 5 punti
  5°: 4 punti
  6°: 3 punti
  7°: 2 punti
  8°: 1 punti
  altre posizioni: 0 punti
  """
  match posizione:
        case 1:
            return 10
        case 2:
            return 8
        case 3:
            return 6
        case 4:
            return 5
        case 5:
            return 4
        case 6:
            return 3
        case 7:
            return 2
        case 8:
            return 1
        case _:
            return 0


def trova_pilota(lista_valori):
  """
  Restituisce statistiche per il pilota richiesto:
  - Numero di vittorie (posizione 1)
  - Numero di podi (posizione 1, 2 o 3)
  - Punti totali nel mondiale (secondo tabella punteggi)
  Se il pilota non è presente, restituisce messaggio di errore
  """
  vittorie = 0
  punti_totali = 0
  podi = 0
  trovato = False

  pilota=input("Nome Pilota: ").strip().lower()

  # Il ciclo for verifica che nel dataset sia presente il nome inserito in input:
  # - Casting del dato Position in intero e assegnazione alla variabile punti
  # - Utilizzo della funzione conversione() per sommare i punti corrispondenti
  # - Ciclo che incrementa i contatori a seconda dei punti corrispondenti es. Vittorie, se punti = 1, Podi, se punti da 1 a 3
  # - Creazione lista "statistiche" che utilizziamo per ritornare i valori

  for riga in lista_valori:
    if riga["Driver"].strip().lower() == pilota:
      trovato = True
      punti = int(riga["Position"])
      punti_totali = punti_totali + conversione(punti)
      if punti == 1:
          vittorie += 1
      if 1 <= punti <= 3:
          podi += 1
  statistiche = [vittorie, podi, punti_totali]

  if not trovato:
    return "Pilota non trovato."

  else:
    return f"Il pilota {pilota.capitalize()}, ha totalizzato {statistiche[2]} punti, {statistiche[0]} vittorie e {statistiche[1]}  podi totali."


def classifica_mondiale_piloti(lista_valori):
  """
  Calcola i punti totali di ogni pilota sul dataset, restituendo un dizionario {pilota: punti}
  """
  classifica ={}

  for riga in lista_valori:
    pilota = riga["Driver"]
    # Modifica stringa in intero e conversione (tramite funzione) nei punteggi corrispondenti
    posizione = int(riga["Position"])
    punti = conversione(posizione)
    if pilota in classifica:
      classifica[pilota] += punti
    else:
      classifica[pilota] = punti

  classifica_ordinata_piloti = dict(sorted(classifica.items(), key=lambda x: x[1], reverse = True))
  return classifica_ordinata_piloti

  # Versione alternativa (più "artigianale") - la mantengo per evidenziare i passi alternativi:
  # lista_piloti_invertita = []
  # for pilota, punti in classifica.items():
  #   lista_piloti_invertita.append((punti, pilota))
  # lista_piloti_invertita.sort(reverse=True)
  # classifica_piloti_ordinata = {}
  # for punti, pilota, in lista_piloti_invertita:
  #   classifica_ordinata[pilota] = punti
  # return classifica_piloti_ordinata


def classifica_mondiale_costruttori(lista_valori):
    """
    Calcola i punti totali per ogni costruttore (team), sommando i punti dei piloti associati,
    e restituisce la classifica ordinata dal primo all'ultimo.

    """
    pilota_team={}

    for riga in lista_valori:
      pilota = riga["Driver"]
      team = riga["Team"]
      pilota_team[pilota] = team

    classifica_piloti = classifica_mondiale_piloti(lista_valori)
    classifica_costruttori = {}

    for pilota, punti in classifica_piloti.items():
      team = pilota_team[pilota]
      if team in classifica_costruttori:
        classifica_costruttori[team] += punti
      else:
        classifica_costruttori[team] = punti

    classifica_ordinata_costruttori = dict(sorted(classifica_costruttori.items(), key=lambda x: x[1], reverse = True))
    return classifica_ordinata_costruttori

    # Versione alternativa (più "artigianale") - vedi sopra(funzione classifica_mondiale_piloti)
    # lista_invertita = []
    # for team, punti in classifica_costruttori.items():
    #   lista_invertita.append((punti, team))
    # lista_invertita.sort(reverse=True)
    # classifica_ordinata = {}
    # for punti, team, in lista_invertita:
    #   classifica_ordinata[team] = punti
    # return classifica_ordinata

```

```python
pilota_da_trovare = (trova_pilota(dati))
classifica_piloti = (classifica_mondiale_piloti(dati))
classifica_costruttori = (classifica_mondiale_costruttori(dati))

print(pilota_da_trovare)
print("classifica piloti 2008: ")
for pilota, punti in classifica_piloti.items():
  print(f"{pilota}: {punti}")
print("Classifica costruttori 2008: ")
for team, punti in classifica_costruttori.items():
  print(f"{team}: {punti}")



# Salvataggio classifica finale piloti su file .txt
with open ("Drivers_Standings_2008.txt", "w")as f:
  f.write("Drivers Standings 2008 Formula 1: ")

  for pilota, punteggio in classifica_piloti.items():
    f.write(f"\n{pilota}: {punteggio}")

```
